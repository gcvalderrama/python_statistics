```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
plt.rcParams["figure.figsize"] = (20,3)
```

```python

numbers = np.random.geometric(p=0.55, size=10000)
sns.kdeplot(numbers)
#sns.lineplot(x=numbers, y=numbers)
plt.axvline( np.median(numbers), 0 , 2, color='red') # median
plt.axvline( np.mean(numbers), 0 , 2, color='yellow') # mean
plt.axvline( st.mode(numbers), 0 , 2, color='purple') # mode
plt.show()
```

```python
median  = np.median(numbers)

max_numbers = list(filter(lambda x: x >= median, numbers))
min_numbers = list(filter(lambda x: x <= median, numbers))

print(len(max_numbers))
print(len(min_numbers))
#print(median)
```
