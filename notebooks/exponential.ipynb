```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
plt.rcParams["figure.figsize"] = (20,3)
```

```python
numbers = np.random.exponential(scale=0.5, size=10000)
```

```python
sns.kdeplot(numbers)
#sns.lineplot(x=numbers, y=numbers)
plt.axvline( np.median(numbers), 0 , 2, color='red') # median
plt.show()
```

```python
print('median {}'.format(np.median(numbers)))
print('mean {}'.format(np.mean(numbers)))
print('mode {}'.format(st.mode(numbers)))
```
